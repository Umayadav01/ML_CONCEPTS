```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x=np.linspace(0,10,100)
y=np.sin(x)
z=np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

# Plotting the data


```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,z)
plt.show()
```

```python
#addid titles 
plt.plot(x,y)
plt.xlabel("angle")
plt.ylabel('sine value')
plt.title("sine wave")
plt.show()
```
