# Seaborn-for  visualization of data
Matplotlib:It is like a blank canvas: It gives you complete control over every detail of your plot, but you have to specify everything explicitly.

Seaborn:IT is like a pre-designed template: It's easier to use and produces attractive plots with less code, especially for statistical visualizations. It's built on top of Matplotlib, so you can still tweak things if needed.









```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#loding bill vs tip dataset
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.relplot(data=tips,x='total_bill',y='tip',col='time',hue='smoker',style='smoker',size='size')
```

```python
sns.set_theme()
sns.relplot(data=tips,x='total_bill',y='tip',col='time',hue='smoker',style='smoker',size='size')
```

```python
#house price pridiction dataset (it is not available in seaborn ,so we will import it from sklearn library)
from sklearn.datasets import fetch_california_housing
housing = fetch_california_housing()
```

```python
#storing it in pandas dataframe
house= pd.DataFrame(housing.data ,columns=housing.feature_names)
house['PRICE']=housing.target
```

```python
house.head()
```

```python
#distribution plot(gives range of data)
sns.distplot(house['PRICE'])
```

```python
#coorelation matrix(negative and positive coorelatin)
corelation=house.corr()
```

```python
#heatmap
plt.figure(figsize=(10,10))
sns.heatmap(corelation,cbar=True,square=True,fmt='.1f',annot=True,annot_kws={'size': 8},cmap='Greens')
```
